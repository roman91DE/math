```python
import numpy as np
```

```python
A = np.array([
    [2, 4],
    [6, 3]
])

A
```

```python
swapper = np.array([
    [0, 1],
    [1, 0]
])
```

```python
swapper @ A
```

```python
A @ swapper
```

```python
ident = np.array([
    [1, 0],
    [0, 1]
])
```

```python
ident @ A
```

```python
A @ ident
```
