```python
import numpy as np
from matplotlib import pyplot as plt
```

# Basic Vectors

```python
a_x = np.array([2,0])
a_y = np.array([0,3])
a = a_x + a_y
```

```python
vectors = np.array((a_x, a_y))
vector_addition = vectors[0] + vectors[1]
vectors = np.append(vectors, vector_addition[None,:], axis=0)
 
tail = [[0, 0, 0], [0, 0, 0]]

fig, ax = plt.subplots(1)
ax.quiver(*tail,
           vectors[:, 0],
           vectors[:, 1],
           scale=1,
           scale_units='xy',
           angles = 'xy',
           color=['g', 'r', 'k'],
         )
 
ax.set_xlim((-1, vectors[:,0].max()+1))
ax.set_ylim((-1, vectors[:,1].max()+1))
# https://danielmuellerkomorowska.com/2020/06/19/plotting-2d-vectors-with-matplotlib/
```
