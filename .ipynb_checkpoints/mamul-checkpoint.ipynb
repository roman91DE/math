```python
A = [
    [3, 2, 1],
    [1, 0, 2]
]

B = [
    [1, 2],
    [0, 1],
    [4, 0]
]
```

```python
def print_matrix(matrix):
    """ print matrix, shape and type for 2d matrices """
    for row in matrix:
        print()
        for col in row:
            print(f"{col:>2}, ", end="")
    print(f"\n\nshape= ({len(matrix)}, {len(matrix[0])})\ntype = {type(matrix[0][0])}")
        

print_matrix(A)
print_matrix(B)
```

```python
def matmul(A, B):
    """ multiplicates two 2dimensional matrices A and B and returns result C (vanilla python) """
    if not len(A) == len(B[0]):   # error checking for dimensions
        raise ValueError
    C = [[0 for _ in range(len(A))] for _ in range(len(B[0]))]
    
```
